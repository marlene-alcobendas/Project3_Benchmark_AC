```python
import pandas as pd

#import all files
df1 = pd.read_csv("North_America_Airbus.csv") 
df2 = pd.read_csv("North_America_Boeing_CFM.csv")
df3 = pd.read_csv("North_America_Boeing_Rest.csv")
df4 = pd.read_csv("EasternEU_Airbus.csv")
df5 = pd.read_csv("EasternEU_Boeing.csv")
df6 = pd.read_csv("West_EU_Airbus.csv")
df7 = pd.read_csv("West_EU_Boeing.csv")
df8 = pd.read_csv("Latin_Airbus.csv")
df9 = pd.read_csv("Latin_Boeing.csv")
iso = pd.read_csv("paises.csv")

#combine all dataframes into one
df_aw = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], ignore_index=True)
df_iso = iso

```

```python
#Clean column names & database entries: PAISES.CSV
df_iso.columns = df_iso.columns.str.strip()

df_iso['name'] = df_iso['name'].astype('string').str.strip().str.lower()

df_iso.head()
```

```python
#Review column required within the exported files
df_iso = df_iso[['name','iso3']]
df_iso.head()
```

```python
#Review column required within the exported files
df_aw = df_aw[['Aircraft Manufacturer', 'Aircraft Type', 'Aircraft Model Variant', 'Engine Manufacturer', 'Engine Type','Engine Model Variant',
       'Aircraft Registration Number','Aircraft Status', 'Aircraft Delivery Date','Operator', 'Operator Country', 'Hours','Cycles', 'Age (Since Delivery)', 'Number of Engines for Operators']]

df_aw.head()
```

```python
#filter the aircraft status to In Service
df_aw = df_aw[df_aw['Aircraft Status'] == 'In Service']

df_aw.head()
```

```python
#Extract AC type from Aircraft Type column, where aircraft manufacturer is Airbus left 4 digits in aircraft type and boeing left 3 digits
df_aw['AC Type'] = df_aw.apply(lambda row: row['Aircraft Type'][:4] if 'Airbus' in row['Aircraft Manufacturer'] else row['Aircraft Type'][:3], axis=1)

df_aw.head()
```

```python
#Review unique values for AC Type
df_aw['AC Type'].unique()
```

```python
#Exclude values "MD-" from column AC type
df_aw = df_aw[~df_aw['AC Type'].str.startswith('MD-', na=False)]
df_aw['AC Type'].unique()
```

```python
#Reorder columns
df_aw = df_aw[['Aircraft Manufacturer', 'AC Type', 'Aircraft Model Variant', 'Engine Manufacturer', 'Engine Type','Engine Model Variant',
       'Aircraft Registration Number','Aircraft Status', 'Aircraft Delivery Date','Operator', 'Operator Country', 'Hours','Cycles', 'Age (Since Delivery)', 'Number of Engines for Operators']]
df_aw
```

```python
#Clean database entries for 'Operator Country':
df_aw['Operator Country'] = df_aw['Operator Country'].astype('string').str.strip().str.lower()

df_aw.head()
```

```python
#Merge df_aw with iso to get the ISO3 code for each country in 'Operator Country' column
df_aw = df_aw.merge(df_iso, how='left', left_on='Operator Country', right_on='name')

df_aw.head()
```

```python
#Reorder columns
df_aw = df_aw = df_aw[['Aircraft Manufacturer', 'AC Type', 'Aircraft Model Variant', 'Engine Manufacturer', 'Engine Type','Engine Model Variant',
       'Aircraft Registration Number','Aircraft Status', 'Aircraft Delivery Date','Operator', 'Operator Country', 'iso3','Hours','Cycles', 'Age (Since Delivery)', 'Number of Engines for Operators']]

df_aw
```

```python
#Convert 'Aircraft Delivery Date' to datetime format and save the year
df_aw['Aircraft Delivery Date'] = pd.to_datetime(df_aw['Aircraft Delivery Date'], errors='coerce', dayfirst=True)
df_aw['Delivery Year'] = df_aw['Aircraft Delivery Date'].dt.year.astype('Int64')

df_aw.head()
```

```python
#Filling missing values in 'Delivery Year' with actual year minus age(since delivery)

current_year = pd.Timestamp.now().year

# 1) Asegura tipos numéricos
df_aw['Age (Since Delivery)'] = pd.to_numeric(df_aw['Age (Since Delivery)'], errors='coerce')
df_aw['Delivery Year'] = pd.to_numeric(df_aw['Delivery Year'], errors='coerce')

# 2) Calcula año candidato solo donde falte (puede dar decimales si la edad no es entera)
deliv_year_calc = current_year - df_aw['Age (Since Delivery)']

# 3) Limpia el resultado: rango razonable y redondea a año
deliv_year_calc = deliv_year_calc.where(deliv_year_calc.between(2000, current_year), other=pd.NA)  # descarta valores raros
deliv_year_calc = deliv_year_calc.round()   # o .floor() si prefieres truncar
deliv_year_calc = deliv_year_calc.astype('Int64')  # ya es seguro convertir

# 4) Redondea y convierte también la columna existente (por si vino como float)
deliv_year = df_aw['Delivery Year'].round().astype('Int64')

# 5) Rellena faltantes con el candidato
df_aw['Delivery Year'] = deliv_year.fillna(deliv_year_calc)

df_aw.head()
```

```python
df_aw['Engine Manufacturer'].unique()

#Replace 'Rolls-Royce Deutshland' with 'Rolls-Royce' in 'Engine Manufacturer' column
df_aw['Engine Manufacturer'] = df_aw['Engine Manufacturer'].replace('Rolls-Royce Deutschland', 'Rolls-Royce')
```

```python
#check unique values again
df_aw['Engine Manufacturer'].unique()
```

```python
#Merge columns Aircraft Manufacturer & Engine Manufaturer in a new column 'ManufacturerAC_Engine'

df_aw['ManufacturerAC_Engine'] = df_aw['Aircraft Manufacturer'] + ' - ' + df_aw['Engine Manufacturer']

df_aw.head()
```

```python
#Group by Engine Manufaturer by Operator Country

#1) Agregación

# a) Recuento DISTINCT por matrícula 
agg = (df_aw .dropna(subset=['ManufacturerAC_Engine', 'iso3', 'Operator Country']) 
       .groupby(['iso3','Operator Country', 'ManufacturerAC_Engine'])['Aircraft Registration Number'] 
       .nunique() 
       .reset_index(name='aircraft_count'))

# Filtro top 10 paises:
top_countries = (agg.groupby('iso3')['aircraft_count'].sum() .nlargest(10).index)

agg = agg[agg['iso3'].isin(top_countries)]


#b) Reagrega por motor
agg_engine = (
    df_aw.dropna(subset=['iso3','Operator Country','Engine Manufacturer','Aircraft Registration Number'])
         .groupby(['iso3','Operator Country','Engine Manufacturer'])['Aircraft Registration Number']
         .nunique().reset_index(name='aircraft_count')
)

# Filtra motores principales
top_engines = (
    agg_engine.groupby('Engine Manufacturer')['aircraft_count'].sum()
              .nlargest(6).index)

agg_engine = agg_engine[agg_engine['Engine Manufacturer'].isin(top_engines)]

#2) Filtrar top K combinaciones ManufacturerAC_Engine

TOP_K = 8

top_combo = (
    agg.groupby('ManufacturerAC_Engine')['aircraft_count']
       .sum()
       .nlargest(TOP_K)
       .index
)

agg_top = agg[agg['ManufacturerAC_Engine'].isin(top_combo)].copy()


#3) Mapa

fig = px.scatter_geo(
    agg_top,
    locations='iso3',
    size='aircraft_count',
    color='ManufacturerAC_Engine',
    facet_col='ManufacturerAC_Engine',
    facet_col_wrap=4,         # cuántos por fila
    size_max=24,
    hover_name='Operator Country',
    projection='natural earth',
    title=f'Distribución por país de las top combinaciones de Manufacturer aircraft & Engine'
)
# Limpiar títulos de faceta
fig.for_each_annotation(lambda a: a.update(text=a.text.split('=')[-1]))
fig.update_layout(showlegend=False, margin=dict(l=10, r=10, t=60, b=10))
fig.show()
```
