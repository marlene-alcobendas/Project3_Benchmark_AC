```python
#Filling missing values in 'Delivery Year' with actual year minus age(since delivery)

current_year = pd.Timestamp.now().year

# 1) Asegura tipos numéricos
df_aw['Age (Since Delivery)'] = pd.to_numeric(df_aw['Age (Since Delivery)'], errors='coerce')
df_aw['Delivery Year'] = pd.to_numeric(df_aw['Delivery Year'], errors='coerce')

# 2) Calcula año candidato solo donde falte (puede dar decimales si la edad no es entera)
deliv_year_calc = current_year - df_aw['Age (Since Delivery)']

# 3) Limpia el resultado: rango razonable y redondea a año
deliv_year_calc = deliv_year_calc.where(deliv_year_calc.between(2000, current_year), other=pd.NA)  # descarta valores raros
deliv_year_calc = deliv_year_calc.round()   # o .floor() si prefieres truncar
deliv_year_calc = deliv_year_calc.astype('Int64')  # ya es seguro convertir

# 4) Redondea y convierte también la columna existente (por si vino como float)
deliv_year = df_aw['Delivery Year'].round().astype('Int64')

# 5) Rellena faltantes con el candidato
df_aw['Delivery Year'] = deliv_year.fillna(deliv_year_calc)

df_aw.head()
```

```python
import re
import pandas as pd

# patrones clave
US_N = re.compile(r"^N\d{1,5}[A-Z]{0,2}$")            # N123, N12345, N123AB
HAS_DASH = re.compile(r"-")

def normalize_registration(reg: str) -> str | None:
    """Normaliza matrículas manteniendo los formatos correctos:
       - USA N-nnnnnLL  -> sin guion
       - Canadá C-F/Gxxx -> con guion (inserta si falta)
       - Prefijos de 1 letra (G,F,D,I,O,Y,M,S,T) -> G-ABCD, etc.
       - Prefijos de 2 chars (EC, PH, OO, 9V, 4X, etc.) -> EC-KGJ, 9V-ABC, etc.
    """
    if pd.isna(reg):
        return None
    s = re.sub(r"\s+", "", str(reg).upper())
    s = re.sub(r"-{2,}", "-", s)  # colapsa guiones múltiples

    # si ya lleva guion y no es una rareza, respeta
    if HAS_DASH.search(s):
        # caso Canadá ya correcto C-Fxxx / C-Gxxx
        if re.match(r"^C-[FG][A-Z]{3}$", s):
            return s
        return s

    # USA: N + 1-5 dígitos + 0-2 letras -> sin guion
    if US_N.match(s):
        return s

    # Canadá sin guion: C F/G + 3 letras -> inserta guion tras C
    if re.match(r"^C[FG][A-Z]{3}$", s):
        return f"C-{s[1:]}"

    # Prefijo de 1 letra + 4 letras (GABCD, FXXXX, DXXXX, IXXXX, etc.)
    if len(s) == 5 and s[0] in set("GFDIOYMSTC") and s[1:].isalpha():
        return f"{s[0]}-{s[1:]}"

    # Prefijo de 2 chars (incluye dígitos: 9V, 4X, 5Y...) + 3–5 alfanum
    m = re.match(r"^([A-Z0-9]{2})([A-Z0-9]{3,5})$", s)
    if m and not s.startswith("N"):  # ya cubrimos N arriba
        return f"{m.group(1)}-{m.group(2)}"

    # si no encaja en ninguna regla, devuelve limpio tal cual
    return s

def norm_reg_series(s: pd.Series) -> pd.Series:
    return s.astype("string").apply(normalize_registration)
```

```python
# 1) Normaliza en tu scrapeo filtrado por años (df_years_filtered)
df_years_filtered["reg_merge"] = norm_reg_series(df_years_filtered["registration"])

asn_by_reg = (
    df_years_filtered
      .dropna(subset=["reg_merge", "date"])
      .groupby("reg_merge", as_index=False)
      .agg(
          asn_engine_events=("date","size"),
          asn_first_event=("date","min"),
          asn_last_event =("date","max"),
      )
)
# 2) Normaliza en tu BBDD df_aw (tu columna se ve como 'Aircraft Registration Number')
df_aw["_reg_merge"] = norm_reg_series(df_aw["Aircraft Registration Number"])

# 3) Merge
df_aw_enriched = (
    df_aw.merge(asn_by_reg, how="left",
                left_on="_reg_merge", right_on="reg_merge")
         .drop(columns=["_reg_merge","reg_merge"])
)
df_aw_enriched["asn_engine_flag"] = df_aw_enriched["asn_engine_events"].fillna(0).gt(0).astype(int)
df_aw_enriched



# 3) Merge
df_aw_enriched = (
    df_aw.merge(asn_by_reg, how="left",
                left_on="_reg_merge", right_on="reg_merge")
         .drop(columns=["_reg_merge","reg_merge"])
)
df_aw_enriched["asn_engine_flag"] = df_aw_enriched["asn_engine_events"].fillna(0).gt(0).astype(int)

df_aw_enriched
```
